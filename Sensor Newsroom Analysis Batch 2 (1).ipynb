```python

#these are the functions to break the rows down into times that can be used in pandas or read by humans. They'll be used 

import time #not sure if I'm using this
import math
def daysSeconds(x,missRatio):
    #first argument is the number of seconds (in this case, the number of rows can be used as a substitute), 2nd argument is the adjustment ratio. (the number of real seconds to recorded data points)
    days= int(math.floor((x*missRatio)/86400))
    seconds=int( (x*missRatio)-(days*86400) ) 
    return days, seconds

def hoursMinutes(y,missRatio):
    hours=int(math.floor((y*missRatio)/3600))
    minutes=int(math.floor(((y*missRatio)-(hours*3600))/60))
    seconds=((y*missRatio)-(hours*3600))-(minutes*60)
    return hours, minutes, seconds

def rebuild(d,h,m,s):
    ss=(d*86400)+(h*3600)+(m*60)+s
    return ss


#this is the function to rebuild a time series. It accepts the dataframe index (to define the length of the new series), the starting time (in %Y-%m-%d %H:%M:%S) and the ratio of recorded seconds to real seconds.
def rewriteTimes(timestamps,firsttime,ratio):
    ratioRemainder=0.0000000
    newTimes=[]
    pos=0
    newTimes.append(pd.to_datetime(firsttime, format="%Y-%m-%d %H:%M:%S")+datetime.timedelta(0,0))
    for timestamp in timestamps:
  #      print 'from original: ',timestamp
 #       print ratioRemainder
        if ratioRemainder<1:
            newtimestamp=newTimes[pos]
            nexttimestamp=newtimestamp+datetime.timedelta(0,1)
            ratioRemainder=ratioRemainder+ratio 
            ratioRemainder=ratioRemainder-1 
            newTimes.append(nexttimestamp)
        else:
            newtimestamp=newTimes[pos]
            nexttimestamp=newtimestamp+datetime.timedelta(0,2)
            ratioRemainder=ratioRemainder+ratio 
            ratioRemainder=ratioRemainder-2 
            newTimes.append(nexttimestamp)
#        print 'generated: ',nexttimestamp,'\n\n'
        pos=pos+1
    del newTimes[-1] #remove the last item so our output list is the same length as our sound and dust lists.
    return newTimes


```

```python
#Set up to clean time in node 1.

def parseTime(t):
    T=[0 for ti in t]
    for i, ti in enumerate(t):
        try:
            T[i]=pd.datetime.strptime(ti,"%Y-%m-%d %H:%M:%S")
        except ValueError:
            T[i]= T[i-1] +datetime.timedelta(0,1)
    return T


def parseDT64(t):
    T=[]
    for i in range(len(t)):
        T.append(pd.datetime.strptime(str(df5.index.values[i])[0:19].replace('T',' '),"%Y-%m-%d %H:%M:%S"))
    return T
```

```python
#prepare to make a new dataframe for the clean, present times
def assembleDF(Times,Dust,Sounds):
    dfSoundInt=[]
    dfDustFloat=[]
    for x in Sounds:
        dfSoundInt.append(int(x))
    for y in Dust:
        dfDustFloat.append(float(y))
    #assemble the datafrom
    T1=Times
    AvgSound = pd.Series(dfSoundInt, index=T1)
    MaxDust = pd.Series(dfDustFloat, index=T1)
    df = pd.DataFrame({'MaxDust':MaxDust,'AvgSound':AvgSound})
    return df
```

```python
#set up a smoothing funtions.
smoothamount=''
#if we use the moving average we can't use times.... have to estimate
def movingAverage(x,seconds):
    window = seconds
    global smoothamount
    smoothamount=str(window)+' seconds'
    i = 0
    out = [0 for _ in range(len(x)-window+1)]
    for xi in x:
        y = np.mean(x[i:i+window])
        out[i]=y
        i = i + 1
        if i+window > len(x):
            print "done."
            break
    
    return out
```

```python
#this is a filter to sort out high and low sound readings.

def filterSound(dfx,soundGate):
    #filters dataframe into low and high groups according to the second variable
    lowPassSounds=[]
    lowPassTimes=[]
    highPassSounds=[]
    highPassTimes=[]
    i=0
    for row_index, row in dfx.iterrows():
        if int(row[1])<soundGate: 
            lowPassSounds.append(int(row[1]))
            lowPassTimes.append(row_index)
        else:
            highPassSounds.append(int(row[1]))
            highPassTimes.append(row_index)
    return lowPassSounds, highPassSounds, lowPassTimes, highPassTimes
```
