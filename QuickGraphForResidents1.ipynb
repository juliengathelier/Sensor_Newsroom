```python
import pandas as pd 
#import seaborn as sns 
import datetime 
import pylab as pb
import numpy as np 



df = pd.DataFrame.from_csv("NODE1_1.TXT")
#df = pd.DataFrame.from_csv("/Users/ferguspitt/Desktop/Node_3_FieldData_06_11.csv")
#df = pd.DataFrame.from_csv("/Users/ferguspitt/Desktop/Node_4_FieldData_06_11.csv")

df.index.name = None

smoothamount=''

#if we use the moving average we can't use times.... have to estimate
def movingAverage(x):
    window = 3600
    global smoothamount
    smoothamount=str(window)+' seconds'
    i = 0
    out = [0 for _ in range(len(x)-window+1)]
    for xi in x:
        y = np.mean(x[i:i+window])
        out[i]=y
        i = i + 1
        if i+window > len(x):
            print "done."
            break
    
    return out


#Only need this if you want to put times in, and you're not using moving averages.

def getTks(Source,cuts):
    cutPoints=[]
    cutLen=len(Source)/cuts
    for k in range(cuts):
        cutPoints.append(str(Source[cutLen*k])[11:16])
    return cutPoints
plainDates=list(df.index)




#Ride through Duplicate Headers

cleanSound=[]
for xx in df["AvgSound"]:
    if xx != 'AvgSound':
        cleanSound.append(int(xx))
    
cleanDust=[]
for yy in df["MaxDust"]:
    if yy != 'MaxDust':
        cleanDust.append(float(yy))

```

```python
        
#pass the clean data sets into the moving average function.
sound= movingAverage(cleanSound)
dust= movingAverage(cleanDust)

figure = figsize(20,10)
pb.plot(sound,linewidth=.5,label='Avg Sound, following '+smoothamount, figure = figure)
pb.axis([0,len(df), 0, np.max(sound)+1])

#put this back if you want the times (or are you too smoothed?)
ax=pb.subplot(1,1,1)
tks=getTks(plainDates,len(ax.get_xticks())-1)
ax.set_xticklabels(tks)
pb.legend()
pb.show()
```

```python
#
#   PLOT THE DUST
#   (inherits the amount of smoothing from the last calculation.


pb.plot(dust, linewidth=.5, label='Avg Dust, following '+smoothamount, figure = figure)

axis([0,len(df), 0, np.max(dust)+1])


#put this back if you want the times (or are you too smoothed?)
ax=pb.subplot(1,1,1)
tks=getTks(plainDates,len(ax.get_xticks())-1)
ax.set_xticklabels(tks)
pb.legend()
pb.show()
```

```python
mm = np.mean(dust)
dustZeroMean = [si-mm for si in dust]


m = np.mean(sound)
soundZeroMean = [si-m for si in sound]
pb.plot(dustZeroMean, label='dust vs mean, following'+smoothamount, alpha=.5, figure = figure)
pb.plot(soundZeroMean, label='sound vs mean, following '+smoothamount, alpha=.5, figure = figure)
pb.legend()
#put this back if you want the times (or are you too smoothed?)
ax=pb.subplot(1,1,1)
tks=getTks(plainDates,len(ax.get_xticks())-1)
ax.set_xticklabels(tks)

pb.show()
```
